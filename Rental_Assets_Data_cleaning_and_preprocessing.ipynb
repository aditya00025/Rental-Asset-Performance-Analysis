```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Rental_Asset_Dataset.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df.columns
```

```python
print("Duplicates:", df.duplicated().sum())
```

```python
df.drop_duplicates(inplace=True)
print("Duplicates:", df.duplicated().sum())
```

```python
df.isnull().sum()
```

```python
df['vendor_name'] = df['vendor_name'].fillna('Unknown Vendor')
df['monthly_rent'] = pd.to_numeric(df['monthly_rent'], errors='coerce')
df['monthly_rent'] = df['monthly_rent'].fillna(df['monthly_rent'].median())

df['city'] = df['city'].astype(str).str.strip()
df['city'] = df['city'].replace(['', 'nan', 'NaN', 'None'], np.nan)
df['city'] = df['city'].fillna('Unknown City')
```

```python
df.isnull().sum()
```

```python
df['vendor_name'] = df['vendor_name'].str.strip().str.title()
print(df['vendor_name'].unique())
```

```python
df['vendor_name'] = df['vendor_name'].replace({
    'Vendorb': 'Vendor B',
    'Vendorb ': 'Vendor B',
    'Vendor A ': 'Vendor A',
    # Add more if needed
})
print(df['vendor_name'].unique())
```

```python
print(df['current_status'].unique())
```

```python
df['current_status'] = df['current_status'].str.strip().str.lower()
print(df['current_status'].unique())
```

```python
df['current_status'] = df['current_status'].replace({
    'in repair': 'In Repair',
    'active': 'Active',
    'retired': 'Retired',
    'returned': 'Returned'
})
print(df['current_status'].unique())
```

```python
df['monthly_revenue_per_asset'] = pd.to_numeric(df['monthly_revenue_per_asset'], errors='coerce')
df['usage_days_(last_30)'] = pd.to_numeric(df['usage_days_(last_30)'], errors='coerce')
```

```python
df['date_of_acquisition'] = pd.to_datetime(df['date_of_acquisition'], errors='coerce')
```

```python
df.head(20)
```

```python
df['monthly_profit'] = df['monthly_revenue_per_asset'] - df['monthly_rent']
df.head(20)
```

```python
df.to_csv('Cleaned_Rental_Asset_Dataset.csv', index=False)
from google.colab import files
files.download("Cleaned_Rental_Asset_Dataset.csv")
```
